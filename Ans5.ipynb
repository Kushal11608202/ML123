## **Importing Libraries**

```
from IPython.display import Image
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# For Half Moon Dataset
from sklearn.datasets import make_moons

#For comparing the k-means clustering, hierarchical clustering (Agglomerative clustering), and DBSCAN in half moon dataset
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from sklearn.cluster import AgglomerativeClustering
```

## **Importing the Moon Dataset**

```
x, y = make_moons(n_samples=170, noise=0.05, random_state=1)
plt.scatter(x[y==0,0],x[y==0,1], edgecolor='black', c='lightblue', marker='o', s=40, label='class 1')
plt.scatter(x[y==1,0],x[y==1,1], edgecolor='black', c='red', marker='s', s=40, label='class 2')
plt.title("Make Moon Dataset :")
plt.legend()
plt.show()
```

## **Comparing the K-Means, Hierarchical and DBSCAN Clusterings in Moon Dataset**

```
figure, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 4))

# K-Means Clustering
km = KMeans(n_clusters=2, random_state=0)
y1 = km.fit_predict(x)

ax1.scatter(x[y1==0,0], x[y1==0,1], edgecolor='black', c='lightblue', marker='^', s=40, label='cluster 1')
ax1.scatter(x[y1==1,0], x[y1==1,1], edgecolor='black', c='red', marker='s', s=40, label='cluster 2')
ax1.set_title('K-means clustering')

# Hierarchical Clustering
ac = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='complete')
y2 = ac.fit_predict(x)

ax2.scatter(x[y2==0,0], x[y2==0,1], edgecolor='black', c='lightblue', marker='^', s=40, label='cluster 1')
ax2.scatter(x[y2==1,0], x[y2==1,1], edgecolor='black',  c='red', marker='s', s=40, label='cluster 2')
ax2.set_title('Hierarchical clustering')

# DBSCAN (Density-based Spatial Clustering of Applications with Noise)
db = DBSCAN(eps=0.2, min_samples=5, metric='euclidean')
y3 = db.fit_predict(x)

ax3.scatter(x[y3==0,0], x[y3==0,1], edgecolor='black', c='lightblue', marker='^', s=40, label='cluster 1')
ax3.scatter(x[y3==1,0], x[y3==1,1], edgecolor='black',  c='red', marker='s', s=40, label='cluster 2')
ax3.set_title('DBSCAN')

plt.legend()
plt.show()
```
