##**Computation Graph :**

It is the basic unit of computation used in TensorFlow. A computation graph consists of nodes and edges. Each node represents an instance of tf.Operation, while each edge represents an instance of tf.Tensor that gets transferred between the nodes.

1. In TensorFlow, model contains a computation graph.
2. Firstly, you need to create the graph with the nodes representing constants, variables, placeholders, and operations.
3. Then provide this graph to the TensorFlow execution engine.
4. The TensorFlow execution engine will find out the first set of nodes that which can execute.
5. The execution of these first set of nodes starts the execution of the next set of nodes that follow the sequence in the computation graph.
6. This program starts the execution with a default graph, Unless the other graph is explicitly specified and a new node gets implicitly added to the default graph.

**Here**

***(a , b, c , d) -> Input Tensors (scalars or constants)***

***(r1 , r2) -> Intermediate result Tensors***

***(z) -> Tensor of the Final Result*** 

```
g = tf.Graph()

with g.as_default():
    a = tf.constant(1, name='a')
    b = tf.constant(2, name='b') 
    c = tf.constant(3, name='c')
    d = tf.constant(4, name='d')
    r1 = a+b+c
    r2 = 3*r1
    z = r2+d
    
with tf.compat.v1.Session(graph=g) as sess:
    print('z =', sess.run(z))
```
